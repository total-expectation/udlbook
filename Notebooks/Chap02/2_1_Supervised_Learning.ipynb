```python
# Math library
import numpy as np
# Plotting library
import matplotlib.pyplot as plt
```

```python
# Create some input / output data
x = np.array([0.03, 0.19, 0.34, 0.46, 0.78, 0.81, 1.08, 1.18, 1.39, 1.60, 1.65, 1.90])
y = np.array([0.67, 0.85, 1.05, 1.0, 1.40, 1.5, 1.3, 1.54, 1.55, 1.68, 1.73, 1.6 ])

print(x)
print(y)
```

```python
# Define 1D linear regression model
def f(x, phi0, phi1):
  # TODO :  Replace this line with the linear regression model (eq 2.4)
  y = phi0 + phi1 * x
  return y
```

```python
# Function to help plot the data
def plot(x, y, phi0, phi1):
    fig,ax = plt.subplots()
    ax.scatter(x,y)
    plt.xlim([0,2.0])
    plt.ylim([0,2.0])
    ax.set_xlabel('Input, $x$')
    ax.set_ylabel('Output, $y$')
    # Draw line
    x_line = np.arange(0,2,0.01)
    y_line = f(x_line, phi0, phi1)
    plt.plot(x_line, y_line,'b-',lw=2)

    plt.show()
```

```python
# Set the intercept and slope as in figure 2.2b
phi0 = 0.4 ; phi1 = 0.2
# Plot the data and the model
plot(x,y,phi0,phi1)
```

```python
# Function to calculate the loss
def compute_loss(x,y,phi0,phi1):
  # TODO Replace this line with the loss calculation (equation 2.5)
  loss = np.sum((f(x, phi0, phi1) - y) ** 2)
  return loss
```

```python
# Compute the loss for our current model
loss = compute_loss(x,y,phi0,phi1)
print(f'Your Loss = {loss:3.2f}, Ground truth =7.07')
```

```python
# Set the intercept and slope as in figure 2.2c
phi0 = 1.60 ; phi1 =-0.8
# Plot the data and the model
plot(x,y,phi0,phi1)
loss = compute_loss(x,y,phi0,phi1)
print(f'Your Loss = {loss:3.2f}, Ground truth =10.28')
```

```python
# TO DO -- Change the parameters manually to fit the model
# First fix phi1 and try changing phi0 until you can't make the loss go down any more
# Then fix phi0 and try changing phi1 until you can't make the loss go down any more
# Repeat this process until you find a set of parameters that fit the model as in figure 2.2d
# You can either do this by hand, or if you want to get fancy, write code to descent automatically in this way
# Start at these values:
phi0 = 1.60 ; phi1 =-0.8





# METHOD 1: GRADIENT DESCENT
# I chose to do it with gradient descent, because it's much more educational than doing trial error by hand. Obviously there are other iterative optimization methods that work as well (like simulated annealing or newton-raphson etc.), but I'm only familiar with GD, so I'll stick to it.
# we need to take the partial derivatives of the loss wrt phi0 and phi1 and update these separately for each step
# dl/dphi0 = 2 * sum(f(x) - y)
# dl/dphi1 = 2 * sum(x * f(x) - y)
# where f(x) = phi0 + phi1x - y
# then according to GD we just update the parameters separately with a learning rate
# phi0 = phi0 - learning_rate * dl/dphi0
# phi1 = phi1 - learning_rate * dl/dphi1
# the iterations and learning rate are hyperparameters, so you have to do them trial and error
learning_rate = 1e-4
iterations = 10000  
# no need for batches, just use the entire data it's fast enough, and we don't care about overfitting because that's what we want in this case (we want the best match on the GIVEN data), whereas in DL in general you want as good as possible on UNSEEN data (so generalization is important)
for i in range(iterations):
    phi0 = phi0 - learning_rate * 2 * np.sum((f(x, phi0, phi1) - y))
    phi1 = phi1 - learning_rate * 2 * np.sum(x * (f(x, phi0, phi1) - y))


plot(x,y,phi0,phi1)
print(f'phi0: {phi0}, phi1: {phi1}')
print(f'Your Loss = {compute_loss(x,y,phi0,phi1):3.2f}')

# METHOD 2: ANALYTICAL CLOSED-FORM SOLUTION
# So instead of using an iterative optimization method, we could just solve this analytically by observing that the zero derivative should give us an optimal answer if the equation is easily solvable. 
# This gives us even a global optimum, because the loss landscape for linear regression with RSS is convex (check that the hessian of the loss is positive semi-definite)
# dL/dph0 = 0
# dL/dphi1 = 0
# where L(x) = np.sum((f(x) - y))^2
# where f(x) = phi0 + phi1x - y
# writing the proof for this is too long, so I will just use the results: https://dafriedman97.github.io/mlbook/content/c1/s1/loss_minimization.html#parameter-estimation
phi1 = np.sum(x * (y - np.mean(y))) / np.sum(x * (x - np.mean(x)))
ph0 = np.mean(y) - phi1 * np.mean(x)
plot(x,y,phi0,phi1)
print(f'phi0: {phi0}, phi1: {phi1}')
print(f'Your Loss = {compute_loss(x,y,phi0,phi1):3.2f}')
```

# Visualizing the loss function

The above process is equivalent to descending coordinate wise on the loss function<br>

Now let's plot that function

```python
# Make a 2D grid of possible phi0 and phi1 values
phi0_mesh, phi1_mesh = np.meshgrid(np.arange(0.0,2.0,0.02), np.arange(-1.0,1.0,0.02))

# Make a 2D array for the losses
all_losses = np.zeros_like(phi1_mesh)
# Run through each 2D combination of phi0, phi1 and compute loss
for indices,temp in np.ndenumerate(phi1_mesh):
    all_losses[indices] = compute_loss(x,y, phi0_mesh[indices], phi1_mesh[indices])
```

```python
# Plot the loss function as a heatmap
fig = plt.figure()
ax = plt.axes()
fig.set_size_inches(7,7)
levels = 256
ax.contourf(phi0_mesh, phi1_mesh, all_losses ,levels)
levels = 40
ax.contour(phi0_mesh, phi1_mesh, all_losses ,levels, colors=['#80808080'])
ax.set_ylim([1,-1])
ax.set_xlabel('Intercept, $\phi_0$')
ax.set_ylabel('Slope, $\phi_1$')

# Plot the position of your best fitting line on the loss function
# It should be close to the minimum
ax.plot(phi0,phi1,'ro')
plt.show()
```
